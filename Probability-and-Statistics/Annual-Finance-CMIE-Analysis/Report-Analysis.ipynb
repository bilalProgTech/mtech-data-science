# Analysis for each year with every different Sectors

```python
anps = ['ANP(Mar-15)', 'ANP(Mar-16)', 'ANP(Mar-17)', 'ANP(Mar-18)', 'ANP(Mar-19)']

sectors = ['Financial', 'Services','Construction', 'FMCG', 
           'Healthcare', 'Chemicals','Engineering', 'Automobile', 
           'Energy', 'Metals', 'Technology', 'Cons Durable', 
           'Textiles', 'Other', 'Communication', 'Diversified']
model_accepted = 0
model_notaccepted = 0
for sector in sectors:
    print('\n')
    print('[[[[[[ SECTOR: ',sector,']]]]]]')
    sec = data[data['SECTOR']==sector]
    for anp in anps:
        print('\n')
        print('--------',anp,'--------')
        X = sec[anp].values.reshape(-1,1)
        for i in range(0,2):
            dep = ''
            if(i==0):
                dep = "CSR SPENT "+anp[3:11]
            if(i==1):
                dep = "CSR UNSPENT "+anp[3:11]
            print('\n')
            print('=====',dep,'=====')
            y = sec[dep].values.reshape(-1,1)
            x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
            regressor = LinearRegression()  
            regressor.fit(x_train, y_train)
            y_pred = regressor.predict(x_test)
            
            #To retrieve the intercept:
            print("Intercept: ",regressor.intercept_)
            
            #For retrieving the slope:
            print("Coefficient: ",regressor.coef_)
            
            #R squared
            n = x_train.shape[0]
            p = x_train.shape[1] - 1
            rsquared = regressor.score(x_train, y_train)
            adj_rsquared = 1 - (1 - rsquared) * ((n - 1) / ( n - p - 1))

            print("R-squared Value: ", rsquared)
            print("Adjusted R-squared Value: ", adj_rsquared)
            if(adj_rsquared>0.80):
                print('$$$ Model is Accepted, hence there is significant relation between these variables $$$')
                model_accepted = model_accepted + 1
            else:
                print('$$$ Model is Not Accepted, hence there is no significant relation between these variables $$$')
                model_notaccepted = model_notaccepted + 1

            print('\n')
            print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))  
            print('Mean Squared Error:', mean_squared_error(y_test, y_pred))  
            print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
            
            df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
            print(df.head(8))
            
            plt.scatter(x_test, y_test,  color='gray')
            plt.plot(x_test, y_pred, color='red', linewidth=2)
            plt.show()

print('Number of Model Accepted: ',model_accepted)
print('Number of Model Not Accepted: ',model_notaccepted)
```
