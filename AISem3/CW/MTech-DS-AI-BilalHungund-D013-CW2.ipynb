```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('food-delivery-time.xlsx')
data.head()
```

```python
test_split = int(len(data)*0.2)

df_train, df_test = data[0:-test_split], data[-test_split:]

df_train.shape, df_test.shape
```

```python
# Traversing colums

for col in data.columns.tolist():
    print(data[col].head(2))
```

```python
# Traversing rows

for i in range(len(data)):
    print('Iteration', i+1, data['Cuisines'][i])
```

```python
def get_value_row(df, value):
    return df[df.eq(value).any(1)]

def get_value_col(df, value):
    return df.iloc[:, np.where(df == value)[1].tolist()[0]]
```

```python
get_value_row(data, 'Biryani, Mughlai')
```

```python
get_value_col(data, '₹100')
```

```python
print('======----Train Data----=======')
for col in df_train.columns.tolist():
    print(df_train[col].head(2))
    
print('======----Test Data----=======')
for col in df_test.columns.tolist():
    print(df_test[col].head(2))
```

```python
get_value_row(df_train, 'Fast Food, North Indian, Rolls, Chinese, Momos, Mughlai')
```

```python
get_value_row(df_test, 'North Indian, Chinese, Fast Food, Mughlai, Italian, Burger, Desserts, Beverages')
```
